# Example of loading in the Bracketed range and plotting it

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load in the data
r = np.load('*path*/MW Inner DM Profile/MW_Bracketed_Range/r.npy')
rho_upper = np.load('*path*/MW Inner DM Profile/MW_Bracketed_Range/predicition_band_upper_bound.npy')
rho_lower = np.load('*path*/MW Inner DM Profile/MW_Bracketed_Range/predicition_band_lower_bound.npy')
```

```python
fig, axs = plt.subplots(1,1,figsize=(14,11))
axs.fill_between(r, rho_upper, rho_lower, color='gray', alpha=0.5)

axs.set_xlim(0.5,10)
axs.set_ylim(10**-(0.72),10**2)
axs.loglog()
axs.tick_params(axis='both', which='major', labelsize=20)
axs.set_xlabel('r [kpc]', fontsize=20)
axs.set_ylabel(r'$\rho$ [GeV/cm$^3$]', fontsize=20)
```
