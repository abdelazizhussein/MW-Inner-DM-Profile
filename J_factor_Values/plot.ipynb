# Example of loading in the J and D factors and plotting them

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load in the J factors
#upper and lower bounds are the J factors for the upper and lower bounds of the MW DM profile

theta = np.rad2deg(np.load('*path*/MW Inner DM Profile/J_factor_Values/theta_array.npy')) #degrees
J_factor_upper_bound = np.load('*path*/MW Inner DM Profile/J_factor_Values/J_MW_prediction_upper.npy') 
J_factor_lower_bound = np.load('*path*/MW Inner DM Profile/J_factor_Values/J_MW_prediction_lower.npy')
```

Note: J factor here is dimensionless, multiply by the appropriate units to get the dimension-full quantitiy

For example, for $[J] = [\rm{GeV}^2/\rm{cm}^5]$ multiply by $\rm\rho_{\odot}^2 r_\odot$ where $\rm\rho_{\odot} = 0.3 \rm{GeV}/\rm{cm}^3$ and $r_\odot = 8 \rm{kpc}$ (Solar Neighborhood values)

```python
fig, axs = plt.subplots(1,1,figsize=(14,11))
axs.plot(theta,J_factor_upper_bound,label='Upper Bound',linewidth=3)
axs.plot(theta,J_factor_lower_bound,label='Lower Bound',linewidth=3)

axs.tick_params(axis='both', which='minor', labelsize=35,size = 20)
axs.tick_params(axis='both', which='major', labelsize=35,size = 20)
axs.legend(frameon=False,fontsize=34)
axs.set_xlabel(r'$\theta^\circ$', fontsize=34)
axs.set_ylabel(r'$J(\theta$)', fontsize=34)

axs.set_yscale('log')

```

```python
#Load in the D factors
#upper and lower bounds are the D factors for the upper and lower bounds of the MW DM profile

theta = np.rad2deg(np.load('*path*/MW Inner DM Profile/J_factor_Values/theta_array.npy')) #degrees
D_factor_upper_bound = np.load('*path*/MW Inner DM Profile/J_factor_Values/D_MW_prediction_upper.npy') 
D_factor_lower_bound = np.load('*path*/MW Inner DM Profile/J_factor_Values/D_MW_prediction_lower.npy')
```
